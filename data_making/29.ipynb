```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 설정
history_1 = {'RMSE': 0.0451, 'R2': 0.6716}
history_2 = {'RMSE': 0.0169, 'R2': 0.9539}

labels = list(history_1.keys())
before_values = [history_1[k] for k in labels]
after_values = [history_2[k] for k in labels]

# 막대 위치 설정
x = np.arange(len(labels))  # 라벨 위치 (0, 1)
width = 0.35               # 막대 너비

fig, ax = plt.subplots(figsize=(6, 6))

# 막대 그래프 생성 (튜닝 전/후를 나란히 배치)
rects1 = ax.bar(x - width/2, before_values, width, label='Before Hyperparameter Tuning', color='#ff9999', edgecolor='gray')
rects2 = ax.bar(x + width/2, after_values, width, label='After Hyperparameter Tuning', color='#66b3ff', edgecolor='gray')

# 차트 정보 설정
ax.set_ylabel('Score / Error Value')
ax.set_title('Model Performance Comparison: Before vs After Tuning', fontsize=14, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2) # 범례를 아래쪽으로 이동
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 각 막대 위에 수치(데이터 라벨)를 표시하는 함수
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),  # 수직으로 5pt 띄움
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.savefig('performance_comparison.png')
plt.show()
```
